```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Connect to the SQLite database file
conn = sqlite3.connect("sales_data.db")
```

```python
# Run SQL query to calculate total quantity and revenue per product
query = '''
SELECT 
    product,
    SUM(quantity) AS total_qty,
    SUM(quantity * price) AS revenue
FROM 
    sales
GROUP BY 
    product
'''
```

```python
# Execute the query and load results into a DataFrame
df = pd.read_sql_query(query, conn)
```

```python
# Print the sales summary
print("ðŸ“¦ Sales Summary:")
print(df)
```

```python

# Plot revenue by product
df.plot(kind='bar', x='product', y='revenue', color='skyblue', title='Revenue by Product')
plt.ylabel("Revenue (INR)")
plt.tight_layout()
plt.savefig("sales_chart.png")  # Save chart as PNG
plt.show()

# Close the database connection
conn.close()
```
